```python
pixel_intersity = [1, 2, 3, 4, 5, 6, 7, 8]
No_of_pixels = []
probability = []
cp = []
cp_20 = []
Floor_Rounding = []
```

```python
for _ in range(int(input())):
  No_of_pixel = int(input())
  No_of_pixels.append(No_of_pixel)
  #probability.append()
```

```python
sum = 0
for _ in No_of_pixels:
  sum+= _
for _ in No_of_pixels:
  probability.append(_/sum)
cp.append(probability[0])
```

```python
for _ in range(1,len(probability)):
  temp = cp[_-1]+probability[_]
  cp.append(temp)


```

```python
from os import cpu_count
for _ in cp:
  cp_20.append(20*_)
```

```python
import math
```

```python
for _ in cp_20:
  Floor_Rounding.append(math.floor(_))
```
